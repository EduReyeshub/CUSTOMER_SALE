Project Overview

This project simulates a real-world data cleaning task where a company needs help preparing a customer database for business use.

The dataset contains customer demographic information and purchasing data.

The objective of this project is to clean and organize the dataset by removing duplicates, handling missing values, and standardizing the data structure.

The final deliverable is a clean Excel file ready to be used for marketing analysis or CRM systems.

Tools used in this project:

Python

Pandas

Google Colab

```
# Import libraries for data analysis and cleaning

import pandas as pd
import numpy as np
```

```
# Load dataset from CSV file
df = pd.read_csv("marketing_campaign.csv", sep=';')

# Display first rows of dataset
df.head()
```

```
# Display dataset structure
df.info()
```

```
# Check missing values

df.isnull().sum()
```

```
# Remove rows with missing income values

df = df.dropna(subset=["Income"])
```

```
# Remove duplicate rows

df = df.drop_duplicates()
```

```
# Calculate total customer spending

df["Total_Spending"] = (
    df["MntWines"]
    + df["MntFruits"]
    + df["MntMeatProducts"]
    + df["MntFishProducts"]
    + df["MntSweetProducts"]
)
```

```
# Select relevant columns for customer database

customer_db = df[[
    "ID",
    "Year_Birth",
    "Education",
    "Marital_Status",
    "Income",
    "Total_Spending",
    "NumWebPurchases",
    "NumStorePurchases"
]]
```

```
# Rename columns to make dataset easier to understand

customer_db = customer_db.rename(columns={
    "Year_Birth": "Birth_Year",
    "NumWebPurchases": "Web_Purchases",
    "NumStorePurchases": "Store_Purchases"
})
```

```
# Sort customers by total spending

customer_db = customer_db.sort_values(by="Total_Spending", ascending=False)
```

```
# Export cleaned dataset to Excel

customer_db.to_excel("clean_customer_database.xlsx", index=False)
```
